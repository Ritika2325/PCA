```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
wine=pd.read_csv("C:/wine.csv")
```

```python
wine.head()
```

```python
wine.tail()
```

```python
# Converting into numpy array
WINE=wine.values
WINE
```

```python
# Normalizing the numerical data
wine_normal=scale(WINE)
```

```python
wine_normal
```

```python
pca=PCA()
pca_values=pca.fit_transform(wine_normal)
```

```python
pca_values
```

```python
pca = PCA(n_components = 6)
pca_values = pca.fit_transform(wine_normal)
```

```python
pca_values
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
#Cummulative Variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
# plot between PCA1 and PCA2 
x = pca_values[:,0:1]
y = pca_values[:,1:2]
plt.scatter(x,y)
```

```python
PCA_KMeans=pca_values[:,0:3]
```

```python
PCA_KMeans
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(PCA_KMeans)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=42)
clusters_new.fit(PCA_KMeans)
```

```python
df_clusters_new = pd.concat([wine.reset_index(drop=True),pd.DataFrame(PCA_KMeans)],axis=1)
df_clusters_new.columns.values[-3:]=['pc1','pc2','pc3']
df_clusters_new['PCAKmeans']=clusters_new.labels_
```

```python
df_clusters_new
```

```python
import seaborn as sns
sns.scatterplot(data=df_clusters_new,x='pc1',y='pc2',hue='PCAKmeans')
```
